```python
import base64
from pathlib import Path

def encode_data(filepath: Path) -> str:
    with Path(filepath).open('rb') as fp:
        bytes_data = base64.b64encode(fp.read())
        encoded_data = bytes_data.decode("utf-8")
        return encoded_data
```

```python
starting_df = pd.DataFrame.from_dict({"name": all_emotions, "score": np.repeat(0.0, 48), "top count": np.repeat(0, 48), "total top score": np.repeat(0.0, 48)})
starting_df
```

```python
dfs = []
n = len(vid_dict)
for frame_num in range(n):
    emotions_dict = vid_dict[frame_num]["emotions"]
    emotions_df = pd.DataFrame.from_dict(emotions_dict).sort_values("name")
    dfs.append(emotions_df)
    starting_df["score"] = starting_df["score"] + emotions_df["score"]
    emotions_df = emotions_df.sort_values("score", ascending=False)
    index = emotions_df.iloc[[0]].index[0]
    starting_df.iloc[index, 2] += 1
    starting_df.iloc[index, 3] += emotions_df.iloc[0, 1]
starting_df["score"] = starting_df["score"] / n
starting_df["mean top score"] = (starting_df["total top score"] / starting_df["top count"]).fillna(0.0)
starting_df
```
